## Casting in C++
### `const_cast<Type>(variable) `
`const_cast`
Its used to change the const or volatile qualifiers of pointers or referances .
where type should be either pointer, reference or pointer to member type

```cpp
int main()
{
    int a=5;
    const int *p=&a;
    int *b=const_cast<int*>(p)
    *b=10;
    cout<<*b<<endl;
    return 0;
}
```
- Result=10;

### `static_cast<Type>(variable)`
Its an implicity conversion in c++

```cpp
int main()
{
    char c='a';
    int data=static_cast<int>(a);
    cout<<data<<endl;
    return 0;
}
```
- Result=10;

### `reinterpret_cast<Type>(variable)`
It is used to cast any pointer type to any pointer type
```cpp
int main()
{
    int a=5;
    int *iptr=&a;
    char *cptr=reinterpret_cast<char*>(&iptr);
    cout<<*cptr<<endl;
    return 0;
}
```

### `dynamic_cast<Type>(variable)`
Its used for safe downcasting at run time.
The `dynamic_cast` to work atleast one virtual function is necessory in the base class
```cpp
class base
{
    public :
    virtual void print()
    {
        cout<<"hey"<<endl;
    }
};

class derived : public base
{
    public :
    virtual void print()override
    {
        cout<<"hey derived"<<endl;
    }
};
int main()
{
    derived d_obj;
    base *b_obj=dynamic_cast<derived*>(&d_obj);
    b_obj->print();
}
```
- Result=`hey derived`

Lambda expression as predicate :
``` cpp
auto n_even = std::count_if(vector.begin(), vector.end(), [](int n) { return (n % 2 == 0); });
```

std::count_if:
This is an STL algorithm that iterates over a given range [first, last) and
counts the elements that satisfy a specific condition (predicate).
The third parameter of std::count_if is a callable object,
which can be a function pointer, a functor, or a lambda function.

```cpp
[](int n) { return (n % 2 == 0); }
```
This is an inline anonymous function (lambda).
It takes one argument (int n), checks whether n is even (n % 2 == 0), 
and returns true if it is even, otherwise false

How the Lambda is Called:
std::count_if will loop through all elements in the range [vector.begin(), vector.end()).
For each element elem in the range, std::count_if will internally call the lambda function:
[](int n) { return (n % 2 == 0); }

passing the current element elem as the argument n.
The lambda function will execute and return a boolean (true or false), 
indicating whether the current element satisfies the condition (is even).
Result Accumulation:
If the lambda function returns true, the internal counter of std::count_if is incremented.
If it returns false, the counter is not changed.
Once the loop completes, std::count_if returns the total count of elements that 
satisfied the condition (i.e., the number of even numbers in this case).


```cpp
//Consider the vector:
std::vector<int> vector = {1, 2, 3, 4, 5};

//The lambda function will be invoked as follows:

For 1: [](int n) { return (n % 2 == 0); } → return (1 % 2 == 0) → false.
For 2: [](int n) { return (n % 2 == 0); } → return (2 % 2 == 0) → true.
For 3: [](int n) { return (n % 2 == 0); } → return (3 % 2 == 0) → false.
For 4: [](int n) { return (n % 2 == 0); } → return (4 % 2 == 0) → true.
For 5: [](int n) { return (n % 2 == 0); } → return (5 % 2 == 0) → false.
//The total count of true results is 2, so n_even = 2.

//lambda local variablaes
auto add_two=[y=2](int x){return x+y}
add_two(2);
```
