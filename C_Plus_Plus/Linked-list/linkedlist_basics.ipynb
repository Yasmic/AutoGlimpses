## single linkedlist

```cpp

// Created by Yasmi on 13/11/24.

#include <iostream>

class Node
{
public:
    int data;
    Node *next;

    explicit Node(int data): data(data), next(nullptr)
    {
    }

    Node(int data, Node *next): data(data), next(nullptr)
    {
    }
};

class singleLinkedList
{
public:
    Node *head;

    singleLinkedList(): head(nullptr)
    {
    }

    void insertAtHead(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            auto *newNode = new Node(data);
            newNode->next = head;
            head = newNode;

        }
    }

    void insertAtTail(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node *temp = head;
            while (temp->next != nullptr)
            {
                temp = temp->next;
            }

            temp->next = new Node(data);
        }
    }

    void insertAtanyPosition(int data, int position)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            int currentPos = 1;
            Node *temp = head;
            Node *previous = nullptr;
            while (temp != nullptr && currentPos < position)
            {

                currentPos++;
                previous = temp;
                temp = temp->next;
            }

            if (currentPos == position)
            {
                Node *newNode = new Node(data);
                newNode->next = temp;
                previous->next = newNode;
            }
        }
    }

    void deleteFromPosition(int position)
    {
        int currentPos = 1;
        Node *temp = head;
        Node *previous = nullptr;
        while (temp != nullptr && currentPos < position)
        {
            currentPos++;
            previous = temp;
            temp = temp->next;
        }

        if (currentPos == position)
        {
            previous->next = temp->next;
        }
    }

    void display() const
    {
        if (head == nullptr)
        {
            std::cout << "List is empty" << std::endl;
        }
        else
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                std::cout << temp->data << " ";
                temp = temp->next;
            }
            std::cout << std::endl;
        }

    }

    ~singleLinkedList()
    {
        if (head != nullptr)
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                const Node *nextNode = temp->next;
                delete temp;
                temp = nextNode;
            }

        }
    }
};

int main()
{
    singleLinkedList obj;
    obj.insertAtHead(12);
    obj.insertAtTail(24);
    obj.insertAtTail(36);
    obj.insertAtTail(48);
    obj.insertAtTail(94);
    std::cout << "list before modification" << std::endl;
    obj.display();
    std::cout << "list after modification" << std::endl;
    obj.insertAtanyPosition(72, 5);
    obj.display();
    std::cout << "list after deleting" << std::endl;
    obj.deleteFromPosition(4);
    obj.display();
    return 0;
}

```
```bash
Output :
list before modification
12 24 36 48 94 
list after modification
12 24 36 48 72 94 
list after deleting
12 24 36 72 94 
```


## double linkedlist

```cpp


#include <iostream>

class Node
{
public:
    int data;
    Node *next;
    Node *prev;

    explicit Node(int data): data(data), next(nullptr),prev(nullptr){}

    Node(int data, Node *next,Node* prev ): data(data), next(nullptr), prev(prev){}
};

class doubleLinkedList
{
public:
    Node *head;

    doubleLinkedList(): head(nullptr){}

    void insertAtHead(int data)
    {
        if(head==nullptr)
        {
            head = new Node(data);
        }
        else
        {
            auto *newNode = new Node(data);
            head->prev = newNode;
            newNode->next = head;
            head = newNode;
            head->prev = nullptr;

        }
    }

    void insertAtTail(int data)
    {
        if(head==nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node * temp=head;
            while(temp->next!=nullptr)
            {
                temp=temp->next;
            }
            Node* newNode = new Node(data);
            newNode->prev=temp;
            temp->next=newNode;
        }
    }

    void insertAtanyPosition(int data,int position)
    {
        if(head==nullptr)
        {
            head = new Node(data);
        }
        else
        {
            int currentPos=1;
            Node * temp=head;
            Node * previous=nullptr;
            while(temp!=nullptr && currentPos<position)
            {
                currentPos++;
                temp=temp->next;
            }

            if(currentPos==position)
            {
                Node * newNode = new Node(data);
                newNode->next=temp;
                temp->prev->next=newNode;
                newNode->prev=temp->prev;
                temp->prev=newNode;
            }
        }
    }

    void deleteFromPosition(int position)
    {
            int currentPos=1;
            Node * temp=head;
            while(temp!=nullptr && currentPos<position)
            {
                currentPos++;
                temp=temp->next;
            }

            if(currentPos==position)
            {
                temp->prev->next=temp->next;
                temp->next->prev=temp->prev;
            }
    }
    void display() const
    {
        if(head==nullptr)
        {
            std::cout<<"List is empty"<<std::endl;
        }
        else
        {
            const Node * temp=head;
            while(temp!=nullptr)
            {
                std::cout<<temp->data<<" ";
                temp=temp->next;
            }
            std::cout<<std::endl;
        }

    }

    ~doubleLinkedList()
    {
        if(head!=nullptr)
        {
           const Node * temp=head;
            while(temp!=nullptr)
            {
                const Node * nextNode=temp->next;
                delete temp;
                temp=nextNode;
            }

        }
    }
};

int main()
{
    doubleLinkedList obj;
    obj.insertAtHead(12);
    obj.insertAtTail(24);
    obj.insertAtHead(6);
    obj.insertAtTail(36);
    obj.insertAtTail(48);
    obj.insertAtTail(94);
    std::cout<<"list before modification" << std::endl;
    obj.display();
    std::cout<<"list after modification" << std::endl;
    obj.insertAtanyPosition(72,6);
    obj.display();
    std::cout<<"list after deletion" << std::endl;
    obj.deleteFromPosition(5);
    obj.display();
    return 0;
}

```

## Reverse single linked list
``` cpp
#include <iostream>
#include<stack>

class Node
{
public:
    int data;
    Node *next;

    explicit Node(int data): data(data), next(nullptr)
    {
    }

    Node(int data, Node *next): data(data), next(nullptr)
    {
    }
};

class singleLinkedList
{
public:
    Node *head;

    singleLinkedList(): head(nullptr)
    {
    }

    void insertAtTail(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node *temp = head;
            while (temp->next != nullptr)
            {
                temp = temp->next;
            }
            temp->next = new Node(data);
        }
    }

    void reverseList()
    {
        Node *prev = nullptr;
        Node *curr = head;
        Node *next = head->next;
        while (next != nullptr)
        {
            curr->next = prev;
            prev = curr;
            curr = next;
            next = next->next;
        }
        curr->next = prev;
        head = curr;
    }

    Node * reverseLeftToRight(Node * &headNode,int left,int right)
    {
        if (left == right) return headNode; // No need to reverse if left == right

        Node* dummy = new Node(0);  // Dummy node to handle edge cases
        dummy->next = headNode;
        Node* prev = dummy;

        // Move `prev` to the node just before `left`
        for (int i = 1; i < left; i++) {
            prev = prev->next;
        }

        Node* start = prev->next;  // The first node in the section to be reversed
        Node* temp = start;
        std::stack<Node*> stk;

        // Push nodes to be reversed onto the stack
        for (int i = left; i <= right; i++) {
            stk.push(temp);
            temp = temp->next;
        }

        // Pop from the stack and reverse the pointers
        Node* nextNode = temp; // The node right after the section to be reversed
        while (!stk.empty()) {
            prev->next = stk.top();
            stk.pop();
            prev = prev->next;
        }

        // Connect the reversed part to the rest of the list
        prev->next = nextNode;

        return dummy->next;
    }
    void display(Node* head) const
    {
        if (head == nullptr)
        {
            std::cout << "List is empty" << std::endl;
        }
        else
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                std::cout << temp->data << " ";
                temp = temp->next;
            }
            std::cout << std::endl;
        }
    }

    ~singleLinkedList()
    {
        if (head != nullptr)
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                const Node *nextNode = temp->next;
                delete temp;
                temp = nextNode;
            }
        }
    }
};

int main()
{
    singleLinkedList obj;
    obj.insertAtTail(24);
    obj.insertAtTail(36);
    obj.insertAtTail(48);
    obj.insertAtTail(72);
    std::cout << "list before reversal" << std::endl;
    obj.display(obj.head);
    obj.reverseList();
    std::cout << "reversed list" << std::endl;
    obj.display(obj.head);
    std::cout <<"reversed left to right"<< std::endl;
    Node* node=obj.reverseLeftToRight(obj.head,1,2);
    obj.display(node);
    return 0;
}
```
``` bash
Output:
list before reversal
24 36 48 72 
reversed list
72 48 36 24 
reversed left to right
48 72 36 24 
```

## Double linkedlist reversal
``` cpp
#include <iostream>

class Node
{
public:
    int data;
    Node *next;
    Node *prev;

    explicit Node(int data): data(data), next(nullptr), prev(nullptr)
    {
    }

    Node(int data, Node *next, Node *prev): data(data), next(nullptr), prev(prev)
    {
    }
};

class doubleLinkedList
{
public:
    Node *head;

    doubleLinkedList(): head(nullptr)
    {
    }

    void insertAtTail(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node *temp = head;
            while (temp->next != nullptr)
            {
                temp = temp->next;
            }
            Node *newNode = new Node(data);
            newNode->prev = temp;
            temp->next = newNode;
        }
    }

    void reverseDoubleList()
    {
        if (head == nullptr)
            return;
        Node *curr = head;
        Node *temp = nullptr;
        while (curr != nullptr)
        {
            temp = curr->prev;
            curr->prev = curr->next;
            curr->next = temp;
            curr = curr->prev;
        }
        if (temp != nullptr)
        {
            head = temp->prev;
        }
    }

    void display() const
    {
        if (head == nullptr)
        {
            std::cout << "List is empty" << std::endl;
        }
        else
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                std::cout << temp->data << " ";
                temp = temp->next;
            }
            std::cout << std::endl;
        }

    }

    ~doubleLinkedList()
    {
        if (head != nullptr)
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                const Node *nextNode = temp->next;
                delete temp;
                temp = nextNode;
            }

        }
    }
};

int main()
{
    doubleLinkedList obj;
    obj.insertAtTail(24);
    obj.insertAtTail(36);
    obj.insertAtTail(48);
    obj.insertAtTail(72);
    std::cout << "linkedlist data " << std::endl;
    obj.display();
    obj.reverseDoubleList();
    std::cout << "reversed list " << std::endl;
    obj.display();
    return 0;
}
```
