## single linkedlist

```cpp

// Created by Yasmi on 13/11/24.

#include <iostream>

class Node
{
public:
    int data;
    Node *next;

    explicit Node(int data): data(data), next(nullptr)
    {
    }

    Node(int data, Node *next): data(data), next(nullptr)
    {
    }
};

class singleLinkedList
{
public:
    Node *head;

    singleLinkedList(): head(nullptr)
    {
    }

    void insertAtHead(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            auto *newNode = new Node(data);
            newNode->next = head;
            head = newNode;

        }
    }

    void insertAtTail(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node *temp = head;
            while (temp->next != nullptr)
            {
                temp = temp->next;
            }

            temp->next = new Node(data);
        }
    }

    void insertAtanyPosition(int data, int position)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            int currentPos = 1;
            Node *temp = head;
            Node *previous = nullptr;
            while (temp != nullptr && currentPos < position)
            {

                currentPos++;
                previous = temp;
                temp = temp->next;
            }

            if (currentPos == position)
            {
                Node *newNode = new Node(data);
                newNode->next = temp;
                previous->next = newNode;
            }
        }
    }

    void deleteFromPosition(int position)
    {
        int currentPos = 1;
        Node *temp = head;
        Node *previous = nullptr;
        while (temp != nullptr && currentPos < position)
        {
            currentPos++;
            previous = temp;
            temp = temp->next;
        }

        if (currentPos == position)
        {
            previous->next = temp->next;
        }
    }

    void display() const
    {
        if (head == nullptr)
        {
            std::cout << "List is empty" << std::endl;
        }
        else
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                std::cout << temp->data << " ";
                temp = temp->next;
            }
            std::cout << std::endl;
        }

    }

    ~singleLinkedList()
    {
        if (head != nullptr)
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                const Node *nextNode = temp->next;
                delete temp;
                temp = nextNode;
            }

        }
    }
};

int main()
{
    singleLinkedList obj;
    obj.insertAtHead(12);
    obj.insertAtTail(24);
    obj.insertAtTail(36);
    obj.insertAtTail(48);
    obj.insertAtTail(94);
    std::cout << "list before modification" << std::endl;
    obj.display();
    std::cout << "list after modification" << std::endl;
    obj.insertAtanyPosition(72, 5);
    obj.display();
    std::cout << "list after deleting" << std::endl;
    obj.deleteFromPosition(4);
    obj.display();
    return 0;
}

```
```bash
Output :
list before modification
12 24 36 48 94 
list after modification
12 24 36 48 72 94 
list after deleting
12 24 36 72 94 
```


## double linkedlist

```cpp


#include <iostream>

class Node
{
public:
    int data;
    Node *next;
    Node *prev;

    explicit Node(int data): data(data), next(nullptr),prev(nullptr){}

    Node(int data, Node *next,Node* prev ): data(data), next(nullptr), prev(prev){}
};

class doubleLinkedList
{
public:
    Node *head;

    doubleLinkedList(): head(nullptr){}

    void insertAtHead(int data)
    {
        if(head==nullptr)
        {
            head = new Node(data);
        }
        else
        {
            auto *newNode = new Node(data);
            head->prev = newNode;
            newNode->next = head;
            head = newNode;
            head->prev = nullptr;

        }
    }

    void insertAtTail(int data)
    {
        if(head==nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node * temp=head;
            while(temp->next!=nullptr)
            {
                temp=temp->next;
            }
            Node* newNode = new Node(data);
            newNode->prev=temp;
            temp->next=newNode;
        }
    }

    void insertAtanyPosition(int data,int position)
    {
        if(head==nullptr)
        {
            head = new Node(data);
        }
        else
        {
            int currentPos=1;
            Node * temp=head;
            Node * previous=nullptr;
            while(temp!=nullptr && currentPos<position)
            {
                currentPos++;
                temp=temp->next;
            }

            if(currentPos==position)
            {
                Node * newNode = new Node(data);
                newNode->next=temp;
                temp->prev->next=newNode;
                newNode->prev=temp->prev;
                temp->prev=newNode;
            }
        }
    }

    void deleteFromPosition(int position)
    {
            int currentPos=1;
            Node * temp=head;
            while(temp!=nullptr && currentPos<position)
            {
                currentPos++;
                temp=temp->next;
            }

            if(currentPos==position)
            {
                temp->prev->next=temp->next;
                temp->next->prev=temp->prev;
            }
    }
    void display() const
    {
        if(head==nullptr)
        {
            std::cout<<"List is empty"<<std::endl;
        }
        else
        {
            const Node * temp=head;
            while(temp!=nullptr)
            {
                std::cout<<temp->data<<" ";
                temp=temp->next;
            }
            std::cout<<std::endl;
        }

    }

    ~doubleLinkedList()
    {
        if(head!=nullptr)
        {
           const Node * temp=head;
            while(temp!=nullptr)
            {
                const Node * nextNode=temp->next;
                delete temp;
                temp=nextNode;
            }

        }
    }
};

int main()
{
    doubleLinkedList obj;
    obj.insertAtHead(12);
    obj.insertAtTail(24);
    obj.insertAtHead(6);
    obj.insertAtTail(36);
    obj.insertAtTail(48);
    obj.insertAtTail(94);
    std::cout<<"list before modification" << std::endl;
    obj.display();
    std::cout<<"list after modification" << std::endl;
    obj.insertAtanyPosition(72,6);
    obj.display();
    std::cout<<"list after deletion" << std::endl;
    obj.deleteFromPosition(5);
    obj.display();
    return 0;
}

```
