## merge list
```cpp
// Created by Yasmi on 13/11/24.

#include <iostream>

class Node
{
public:
    int data;
    Node *next;

    explicit Node(int data): data(data), next(nullptr)
    {
    }

    Node(int data, Node *next): data(data), next(nullptr)
    {
    }
};

class singleLinkedList
{
public:
    Node *head;

    singleLinkedList(): head(nullptr)
    {
    }

    void insertAtTail(int data)
    {
        if (head == nullptr)
        {
            head = new Node(data);
        }
        else
        {
            Node *temp = head;
            while (temp->next != nullptr)
            {
                temp = temp->next;
            }

            temp->next = new Node(data);
        }
    }

    Node *mergeLists(Node *&list1, Node *&list2)
    {
        if (list1 == nullptr)
            return list2;
        if (list2 == nullptr)
            return list1;

        Node *last = nullptr;
        Node *third = nullptr;
        while (list1 != nullptr && list2 != nullptr)
        {
            if (list1->data < list2->data)
            {
                if (third == nullptr)
                {
                    last = third = list1;
                }
                else
                {
                    last->next = list1;
                    last = last->next;
                }
                list1 = list1->next;

            }
            else
            {
                if (third == nullptr)
                {
                    last = third = list2;
                }
                else
                {
                    last->next = list2;
                    last = last->next;
                }
                list2 = list2->next;
            }
        }
        if (list1 != nullptr)
        {
            last->next = list1;
            list1 = nullptr;
        }
        else if (list2 != nullptr)
        {
            last->next = list2;
            list2 = nullptr;
        }
        return third;
    }


    void display(Node *head) const
    {
        if (head == nullptr)
        {
            std::cout << "List is empty" << std::endl;
        }
        else
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                std::cout << temp->data << " ";
                temp = temp->next;
            }
            std::cout << std::endl;
        }

    }

    ~singleLinkedList()
    {
        if (head != nullptr)
        {
            const Node *temp = head;
            while (temp != nullptr)
            {
                const Node *nextNode = temp->next;
                delete temp; // Frees the current node
                temp = nextNode; // Moves to the next node
            }
        }
    }
};

int main()
{
    singleLinkedList obj;
    obj.insertAtTail(1);
    obj.insertAtTail(3);
    obj.insertAtTail(5);
    std::cout << "list1" << std::endl;
    obj.display(obj.head);

    singleLinkedList obj1;
    obj1.insertAtTail(2);
    obj1.insertAtTail(4);
    obj1.insertAtTail(6);
    obj1.insertAtTail(8);
    std::cout << "list2" << std::endl;
    obj1.display(obj1.head);

    singleLinkedList obj2;
    Node *node = obj2.mergeLists(obj.head, obj1.head);
    std::cout << "merged list" << std::endl;
    obj2.display(node);
    return 0;
}

```
